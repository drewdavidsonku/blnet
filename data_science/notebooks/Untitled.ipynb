```python
import pandas as pd
import numpy as np
import os
import json
```

```python
os.listdir('../data')
```

```python
raw_data = pd.read_json('../data/isbn_data.json')
raw_data
```

```python
#remove all source tags
src_tags = []
for tag in raw_data.columns:
    if "_source" in tag:
        src_tags.append(tag)
        
print(src_tags)
```

```python
just_data = raw_data.drop(src_tags, axis=1)
just_data = just_data.fillna(value=False)
just_data
```

```python
#change download to true or false
just_data['download'] = np.where(just_data['download'] != False, 1, 0)
just_data
```
